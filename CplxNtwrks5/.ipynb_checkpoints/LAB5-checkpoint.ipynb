```python
def sir(beta, r, N, days):
    N=N
    I=[1/N]
    S=[1.0-I[0]]
    R=[0]
    T=[0]
    
    for t in range(days):
        s=S[t]-I[t]*beta*S[t]
        i=I[t]+I[t]*beta*S[t]-I[t]*r
        r=R[t]+I[t]*r
        
        S.append(s)
        I.append(i)
        R.append(r)
        T.append(t+1)
        
    plt.plot(T,S, 'green', label='Susceptible')
    plt.plot(T,I, 'red', label='Infected')
    plt.plot(T,R, 'black', label='Recovered')
    plt.xlabel('Days')
    plt.ylabel('Fraction')
    
    
    
```

```python
#task 2

def infection_simulation(G: nx.Graph, p: float, source: int):
    #setting all nodes as susceptible
    states={node:0 for node in G.nodes()}
    nx.set_node_attributes(G,states,"state")
    source=list(G.nodes())[source]
    G.nodes[source]['state']=1
    infects=[]
    step=0
    #step_graphs=[] #list of graphs at each MCS
    
    while True: #loop if there are still infected people
        #step_graphs.append(G)
        curr_states=nx.get_node_attributes(G, "state")
        
        infected = [x for x,y in G.nodes(data=True) if y['state']==1] # take all the infected nodes
        infects.append((step,len(infected)))
        for i in infected:
            
            targets=nx.neighbors(G,i) #targets that the infection can spread to
            for t in targets:

                prob=np.random.uniform()
                if prob<p and G.nodes[t]['state']==0:
                    G.nodes[t]['state']=1
                    
            G.nodes[i]['state']=-1

        if 1 not in curr_states.values(): # if no infected remain break the loop
            break
        step+=1
    return infects,G
        
            
        

```

```python

def sim_task_d(ps,G):
    fracs=[]
    clear_times=[]
    max_times=[]
    for p in ps:
        avg_frac=0 #fraction of infected
        avg_clear=0
        avg_max=0
        for i in range(500):
            src=np.random.choice(range(100))   
            sim=infection_simulation(G,p,src)
            sim_path=sim[0]
            sim_graph=sim[1]

            #fraction of infected population
            recovered=nx.get_node_attributes(sim_graph, "state")
            frac=-sum([int(i) for i in list(recovered.values())])/len(G)
            avg_frac+=frac
            ######

            #time to end the infection
            t_last=sim_path[-1][0]
            avg_clear+=t_last
            ######

            #time to largest ammount of infected
            ts=[x[0] for x in sim_path]
            infs=[x[1] for x in sim_path]
            ind=infs.index(max(infs)) #index of maximum infections
            t_max=ts[ind]
            avg_max+=t_max
            
        fracs.append(avg_frac/500)
        clear_times.append(avg_clear/500)
        max_times.append(avg_max/500)
    return fracs,clear_times,max_times
```

```python
ps=np.arange(0.05,1.05,0.05)

sqr_lat=sim_task_d(ps,A1)
rnd_graph=sim_task_d(ps,B1)
watts_strogatz=sim_task_d(ps,C1)
barabasi_albert=sim_task_d(ps,D1)
```

```python

def draw_sims(G: nx.Graph, p: float, source: int, name: str):
    
    #setting all nodes as susceptible
    states={node:0 for node in G.nodes()}
    nx.set_node_attributes(G,states,"state")
    source=list(G.nodes())[source]
    G.nodes[source]['state']=1
    infects=[]
    step=0
    seed = 10
    pos = nx.random_layout(G, seed=seed)
    
    while True: #loop if there are still infected people
        
        curr_states=nx.get_node_attributes(G, "state")
        color_map=nx.get_node_attributes(G, "state")
        for key in color_map:
            if color_map[key]==0:
                color_map[key]='blue'
            elif color_map[key]==1:
                color_map[key]='red'
            elif color_map[key]==-1:
                color_map[key]='green'
        node_colors=[color_map.get(node) for node in G.nodes()]
        nx.draw(G,pos=pos, node_color=node_colors)
        plt.title(f"Infection run for {name} graph")
        plt.savefig(f'LAB5/images/frame_{step}.png', 
                transparent = False,  
                facecolor = 'white'
            )
        
        infected = [x for x,y in G.nodes(data=True) if y['state']==1] # take all the infected nodes
        infects.append((step,len(infected)))
        for i in infected:
            
            targets=nx.neighbors(G,i) #targets that the infection can spread to
            for t in targets:

                prob=np.random.uniform()
                if prob<p and G.nodes[t]['state']==0:
                    G.nodes[t]['state']=1
                    
            G.nodes[i]['state']=-1

        if 1 not in curr_states.values(): # if no infected remain break the loop
            break
        step+=1
    frames = []
    for t in range(step):
        image = imageio.v2.imread(f'LAB5/images/frame_{t}.png')
        frames.append(image)

    imageio.mimsave(f'LAB5/{name}_animated.gif', # output gif
                    frames,
                    format='GIF',          # array of input frames
                    duration=1000)         # optional: frames per second
    
```

```python
draw_sims(A2, 0.9, 0, "Square lattice")
draw_sims(B2, 0.9, 0, "Random")
draw_sims(C2, 0.9, 0, "Watts Strogatz")
draw_sims(D2, 0.9, 0, "Barabasi Albert")
```

```python
def infection_simulation_async(G: nx.Graph, p: float, source: int):
    """
    
    
    
    
    """
    #setting all nodes as susceptible
    states={node:0 for node in G.nodes()}
    nx.set_node_attributes(G,states,"state")
    source=list(G.nodes())[source]
    G.nodes[source]['state']=1
    infects=[]
    step=0
    
    
    while True: #loop if there are still infected people
        #step_graphs.append(G)
        curr_states=nx.get_node_attributes(G, "state")
        
        infected = [x for x,y in G.nodes(data=True) if y['state']==1] # take all the infected nodes
        infects.append((step,len(infected)))
        
        rand_nodes=np.random.choice(range(len(G)),len(G))
        for i in rand_nodes:
            x=list(G.nodes())[i]
            if G.nodes[x]['state']==1:
                targets=nx.neighbors(G,x) #targets that the infection can spread to
                for t in targets:

                    prob=np.random.uniform()
                    if prob<p and G.nodes[t]['state']==0:
                        G.nodes[t]['state']=1
                        
                G.nodes[x]['state']=-1

        if 1 not in curr_states.values(): # if no infected remain break the loop
            break
        step+=1
    return infects,G
```

```python
def sim_task_d_async(ps,G):
    fracs=[]
    clear_times=[]
    max_times=[]
    for p in ps:
        avg_frac=0 #fraction of infected
        avg_clear=0
        avg_max=0
        for i in range(100):
            src=np.random.choice(range(100))   
            sim=infection_simulation_async(G,p,src)
            sim_path=sim[0]
            sim_graph=sim[1]

            #fraction of infected population
            recovered=nx.get_node_attributes(sim_graph, "state")
            frac=-sum([int(i) for i in list(recovered.values())])/len(G)
            avg_frac+=frac
            ######

            #time to end the infection
            t_last=sim_path[-1][0]
            avg_clear+=t_last
            ######

            #time to largest ammount of infected
            ts=[x[0] for x in sim_path]
            infs=[x[1] for x in sim_path]
            ind=infs.index(max(infs)) #index of maximum infections
            t_max=ts[ind]
            avg_max+=t_max
            
        fracs.append(avg_frac/100)
        clear_times.append(avg_clear/100)
        max_times.append(avg_max/100)
    return fracs,clear_times,max_times
```

```python
ps=np.arange(0.05,1.05,0.05)

sqr_lat_async=sim_task_d_async(ps,A1)
rnd_graph_async=sim_task_d_async(ps,B1)
watts_strogatz_async=sim_task_d_async(ps,C1)
barabasi_albert_async=sim_task_d_async(ps,D1)
```

```python
def draw_sims_async(G: nx.Graph, p: float, source: int, name: str):
    
    #setting all nodes as susceptible
    states={node:0 for node in G.nodes()}
    nx.set_node_attributes(G,states,"state")
    source=list(G.nodes())[source]
    G.nodes[source]['state']=1
    infects=[]
    step=0
    seed = 10
    pos = nx.random_layout(G, seed=seed)
    
    while True: #loop if there are still infected people
        
        curr_states=nx.get_node_attributes(G, "state")
        color_map=nx.get_node_attributes(G, "state")
        for key in color_map:
            if color_map[key]==0:
                color_map[key]='blue'
            elif color_map[key]==1:
                color_map[key]='red'
            elif color_map[key]==-1:
                color_map[key]='green'
        node_colors=[color_map.get(node) for node in G.nodes()]
        nx.draw(G,pos=pos, node_color=node_colors)
        plt.title(f"Infection run for {name} graph")
        plt.savefig(f'LAB5/images/frame_{step}.png', 
                transparent = False,  
                facecolor = 'white'
            )
        
        infected = [x for x,y in G.nodes(data=True) if y['state']==1] # take all the infected nodes
        infects.append((step,len(infected)))

        rand_nodes=np.random.choice(range(len(G)),len(G))
        for i in rand_nodes:
            x=list(G.nodes())[i]
            if G.nodes[x]['state']==1:
            
                targets=nx.neighbors(G,x) #targets that the infection can spread to
                for t in targets:

                    prob=np.random.uniform()
                    if prob<p and G.nodes[t]['state']==0:
                        G.nodes[t]['state']=1
                        
                G.nodes[x]['state']=-1

        if 1 not in curr_states.values(): # if no infected remain break the loop
            break
        step+=1
    frames = []
    for t in range(step):
        image = imageio.v2.imread(f'LAB5/images/frame_{t}.png')
        frames.append(image)

    imageio.mimsave(f'LAB5/{name}_async_animated.gif', # output gif
                    frames,
                    format='GIF',          # array of input frames
                    duration=1000)         # optional: frames per second
```

```python
draw_sims_async(A2, 0.9, 0, "Square lattice")
draw_sims_async(B2, 0.9, 0, "Random")
draw_sims_async(C2, 0.9, 0, "Watts Strogatz")
draw_sims_async(D2, 0.9, 0, "Barabasi Albert")
```
